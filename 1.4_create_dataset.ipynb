```python
import pandas as pd
```

```python
oc = pd.read_csv('tmp/opencorpora.csv.gz', compression='gzip')
criminal_code = pd.read_csv('tmp/criminal_code.csv')
criminal_court_orders = pd.read_csv('tmp/criminal_code_court_orders.csv.gz', compression='gzip')
civil_code = pd.read_csv('tmp/vectors/civil_code.csv.gz', compression='gzip')
civil_court_orders = pd.read_csv('tmp/civil_code_court_orders.csv.gz', compression='gzip', index_col=0)

oc['source'] = 'opencorpora'
criminal_code['source'] = 'criminal_code'
criminal_court_orders['source'] = 'criminal_court_orders'
civil_code['source'] = 'civil_code'
civil_court_orders['source'] = 'civil_court_orders'

civil_court_orders['name'] = civil_court_orders['Номер дела'].map(lambda x: f"Гражданское дело №{x}")
criminal_code['name'] = criminal_code.apply(lambda x: "Ст.{} {}".format(x['article_number'], x['article_name']), axis=1)
civil_code['name'] = civil_code.apply(lambda x: "Ст.{} {}".format(x['article_number'], x['article_name']), axis=1)
```

```python
df = pd.concat([
    oc.reset_index()[['title', 'text', 'lemmas', 'source']].rename(columns={'title':'name'}), 
    criminal_code.reset_index()[['name', 'article', 'article_number', 'lemmas', 'source']].rename(columns={"article": "text", 'article_number': "number"}),
    criminal_court_orders.reset_index()[['title', 'case', 'lemmas',  'source', 'labels']].rename(columns={'title':'name', 'case': 'text'}),
    civil_code.reset_index()[['name', 'article', 'article_number', 'lemmas', 'source']].rename(columns={"article": "text", "article_number": "number"}),
    civil_court_orders.reset_index()[['name', 'case', 'lemmas', 'source']].rename(columns={'case':'text'}),
]).reset_index(drop=True)
```

```python
df.sample(10)
```

```python
df.to_csv("DATASET.csv.gz", compression='gzip', index=False)
```
